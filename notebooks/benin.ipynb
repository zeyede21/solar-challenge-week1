```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore

# Display settings
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")
```

```python
import pandas as pd

# Use full path with raw string notation (r"...")
df = pd.read_csv(r"D:\10Academy1\solar-challenge-week1\data\benin-malanville.csv")

# Display shape and columns
print(f"Shape: {df.shape}")
print(df.columns)
```

```python
df.describe(include='all')
```

```python
df.isna().mean().sort_values(ascending=False) * 100
```

```python
df.dtypes
```

```python
columns_to_check = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']
z_scores = df[columns_to_check].apply(zscore)

# Identify rows where any column's z-score is greater than 3 (outlier)
outliers = (abs(z_scores) > 3).any(axis=1)

# Report number of outliers
print(f"Outliers detected: {outliers.sum()} rows")

# Drop outliers
df_clean = df[~outliers].copy()
```

```python
df_clean.fillna(df_clean.median(numeric_only=True), inplace=True)
```

```python
df_clean.isna().sum()
```

```python
df_clean['Timestamp'] = pd.to_datetime(df_clean['Timestamp'])
df_clean.set_index('Timestamp', inplace=True)
```

```python
df_clean[['GHI', 'DNI', 'DHI']].plot(figsize=(15, 5), title="Solar Irradiance")
plt.ylabel("Irradiance (W/m^2)")
plt.show()
```

```python
df_clean[['Tamb', 'WS']].plot(figsize=(15, 5), title="Ambient Temp and Wind Speed")
plt.ylabel("Value")
plt.show()
```

```python
df_clean.reset_index().to_csv('../data/benin_clean.csv', index=False)
```
